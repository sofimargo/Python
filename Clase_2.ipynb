## Elif
Ahora hagamos un programa en el que preguntamos la calificación del 0 al 100 y queremos saber su equivalente en el sistema de evaluación estadounidense. Para simplicidad, basemonos en esto:
- A+ = 100
- A = 90 - 99
- B = 80 - 89
- C = 70 - 79
- D = 60 - 69
- F = < 60

```python
cal = 80
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else: # Notemos como al ya no haber una condición el else cubre todo lo demás
  print("F")
```

## Operadores Lógicos con Condicionales

```python
# and, or, not
pan = True
queso = True
mayonesa = True
jamón = False
print(pan and queso and mayonesa and jamón)
if not (pan and queso and mayonesa and jamón):
  print("No te puedes hacer el sandwich")
```

```python
cal = 80
if cal >= 0 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else: # Notemos como al ya no haber una condición el else cubre todo lo demás
    print("F")
else:
  print("Error: La calificación debe estar entre 0 y 100")
```

```python
# El modulo (%)
# El módulo es un operador matemático que devuelve el residuo de una división
#10 % 3 da 1, porque 10 vidivido entre 3 cabe 3 veces y sobra 1
#7 % 2 da 1, porque no es divisible de forma exacta entre 2
# 2 % 2 -> 0
num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El número {num} es multiplo de 3 o de 5")
# saber si el número es par o impar usando not
num = 6
if not num % 2 == 0:
  print(f"El úmero {num} es impar")
```

## CICLO FOR
Los llamados ciclos for, nos ayuda a realizar acciones durante una cantidad determinada de veces

```python
# Imprimir los números del 0 al 4
for i in range(5): # Toma 5 como el número final pero inicia siempre en 0
  print(i)
```

```python
for num in range(2,11):
  print(num)
```

```python
# También existe el range(inicio,fin, paso)
for i in range(2,21,2):
  print(i)
```

```python
cadena = "Python"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

##While

```python
suma = 0
numeros = int(input("Ingresa el número para sumar o 0 para salir"))

while numeros != 0: # != es equivalente a distinto de ...
  suma += numeros
  numeros = int(input("Ingresa el número para sumar o 0 para salir"))
print(f"Tu suma es: {suma}")
```

```python
# Usar el while como for
numero = 0
while numero <= 12:
  print(numero)
  numero += 1
```
