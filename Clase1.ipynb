### Operaciones Matemáticas

```python
# 4to Nivel de importancia
#Suma Ordinaria
print(3+4)
variable1,variable2 = 20,40
print(variable1 + variable2)

#Resta Ordinaria
print(2-5)

#Declaración de variables
v1 = 10.4
v2 = 15
resultado = v2 - v1
print(resultado)
```

```python
# 3er Nivel de Importancia
#Multiplicación Ordinaria
print("Multiplicación Ordinaria")
m1,m2 = 2,3.14
print(m1*m2)

#División Ordinaria
print("-------------")
print("División Ordinaria")
m1,m2,m3,m4,m5,m6 = 2,5,6,5,4,5
print(m1/m2)
print(m3/m4)
print(m5/m6)
# División de Piso
print("------------")
print("División de Piso")
print(m1//m2) #se redondea hacia abajo
print(m3//m4)
print(m5//m6)

print(6/2*3)
```

```python
# Potencias
print("Potencias")
print(2**3)
P1,P2 = 7,2
print(P1**P2)
```
