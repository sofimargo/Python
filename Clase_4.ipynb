## Funciones
En este curso hemos usado algunas funciones integradas por python, pero también nosotros podemos definir funciones, siendo estas un bloque de código reutilizable que podemos mandar a llamar varias veces en el código

```python
def mensaje():
  print("Hola!!!")

mensaje()
```

## Funciones que llaman a otras Funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"

def despedirse(nombre):
  return f"Adios, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + " " + despedida
print(mensaje_completo("Ana"))
```

```python
# Ahora supongamos que tenemos una heladería
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "fresa"]
  print("Sabor disponible", ",", sabores)
  # lower() hace minusculas las palabras
  # strip() elimina los espacios tanto antes como después
  sabor = input("Elige un sabor:").strip().lower()
# si sabor NO está en sabores
  while sabor not in sabores:
    sabor = input("Elige un sabor (nuevamente): ") .strip().lower()
  return sabor
def pedir_cantidad():
  cantidad = int(input("Cuantas bolas de nieve?: "))
  while cantidad <= 0:
    cantidad = int(input("Cuantas bolas de nieve? (ingresa un valor valido):"))
  return cantidad
# Función Madre (llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

total_pedidos = int(input("Cuantos clientes atenderás?: "))
for i in range(total_pedidos):
  pedir_helado()
```

# Funciones Lambda
Una función lambda es una función acortada para escribir una función de esta forma:
nombreFunción = lambda parametros: acción
La diferencia entre las funciones lambda y las funciones normales es que suelen ser limitadas

```python
sumar = lambda x,y : x+y
print(sumar(3,10))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1-desc)),(precio * desc))
nuevoPrecio = descuentoLambda(0.4,500)
print("El precio a pagar es", nuevoPrecio[0], ", el descuento que se hizo fue de:", nuevoPrecio[1])
print(type(nuevoPrecio))
```

```python
# Calculadora -> suma, resta, producto, división y potencia
def suma(a,b):
  return a + b
def resta(a,b):
  return a - b
def producto(a,b):
  return a * b
def division(a,b):
  if b == 0:
    return "Error: no se puede dividir entre 0"
  return a/b
def potencia(a,b):
  return a**b

def menu():
  print("-------- MI CALCULADORA -------")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. División")
  print("5. Potencia")
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("seleccione una opción: "))
  except ValueError:
    print("Ingrese un número valido")
  if op == 6:
    print("SALIENDO...")
    break
  if op < 1 or op > 6:
      print("Opción invalida")
      continue

  try:
    a = float(input("Ingrese el primero número: "))
    b = float(input("Ingrese el segundo número: "))
  except ValueError:
    print("Error: Ingrese un número valido")
  if op == 1:
    print("Resultado:", suma(a,b))
  elif op == 2:
    print("Resultado:", resta(a,b))
  elif op == 3:
    print("Resultado:", producto(a,b))
  elif op == 4:
    print("Resultado:", division(a,b))
  elif op == 5:
    print("Resultado:", potencia(a,b))
  else:
    print("Opción invalida")
```
